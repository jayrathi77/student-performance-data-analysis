```python
import pandas as pd

df = pd.read_csv("dataset/Student Performance Analysis Dataset.csv")

df.head()
```

```python
print("shape of dataset:",df.shape)

print("column names:",df.columns)

print("\nData Types:")
print(df.dtypes)

print("\nDataset Information:")
df.info()
```

```python
import numpy as np

numeric_df = df.select_dtypes(include=np.number)

numeric_df.head()
```

```python
mean = numeric_df.mean()

print("Mean of Numerical Columns:")
print(mean)
```

```python
median = numeric_df.median()

print("Median of Numerical Columns:")
print(median)
```

```python
mode = numeric_df.mode()

print("Mode of Numerical Columns:")
print(mode)
```

```python
std = numeric_df.std()

print("Standard Deviation:")
print(std)
```

```python
# Select 20% random sample from the dataset
sample = df.sample(frac=0.20, random_state=42)

print("Sample Dataset:")
sample.head()
```

```python
sample_numeric = sample.select_dtypes(include=np.number)

sample_numeric.head()
```

```python


print("Mean")
print(sample_numeric.mean())

print("\nMedian")
print(sample_numeric.median())

print("\nMode")
print(sample_numeric.mode())

print("\nStandard Deviation")
print(sample_numeric.std())
```

```python
original_frequency = df["math_score"].value_counts().sort_index()

print("Original Dataset Frequency Count")
print(original_frequency)
```

```python
# Frequency count of Age column in the original dataset

original_frequency = df["age"].value_counts().sort_index()

print("Original Dataset Frequency Count")
print(original_frequency)
```

```python
# Frequency count of Age column in the sample dataset

sample_frequency = sample["age"].value_counts().sort_index()

print("Sample Dataset Frequency Count")
print(sample_frequency)
```

```python
comparison = pd.DataFrame({
    "Original Frequency": original_frequency,
    "Sample Frequency": sample_frequency
}).fillna(0)

comparison
```

```python
import matplotlib.pyplot as plt

# Plot comparison graph
comparison.plot(
    kind="bar",
    figsize=(10,6),
    width=0.8
)

plt.title("Original Dataset vs Sample Dataset Frequency Comparison", fontsize=14)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.legend(["Original Dataset", "Sample Dataset"])

plt.tight_layout()
plt.show()
```
